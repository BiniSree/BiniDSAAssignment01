```python
#finding the outliers using Z score method
#Outlier in Age
```

```python
age_data= data['Age']
z_score= (age_data-age_data.mean())/age_data.std()
z_score.mean()
```

```python
threshold= 3
outliers= data[abs(z_score)>threshold]
outliers
```

```python
#Outlier in Fare
```

```python
fare_data= data['Fare']
z_score= (fare_data-fare_data.mean())/fare_data.std()
z_score.mean()
```

```python
threshold= 7
outliers
```

```python
#Outlier in SibSp
```

```python
sib_data= data['SibSp']
z_score= (sib_data-sib_data.mean())/sib_data.std()
z_score.mean()
```

```python
threshold= 6
outliers= data[abs(z_score)>threshold]
outliers
```

```python
#Finding the outliers using IQR method
```

```python
Q1= data.quantile(0.25)
Q3= data.quantile(0.75)
IQR= Q3-Q1
```

```python
threshold= 1.5
```

```python
outliers= data[(data < Q1 - threshold * IQR) | (data > Q3 + threshold * IQR)]
outliers
```
